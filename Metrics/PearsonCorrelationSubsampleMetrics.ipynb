```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Means = pd.read_csv("Mean_Pearsons_s01.csv")
SDs = pd.read_csv("SD_Pearsons_s01.csv")
Means
```

```python
sample_sizes = np.array(SDs["Unnamed: 0"])
SDs = SDs.drop(columns=['Unnamed: 0', 'PCA'])
Means = Means.drop(columns=['Unnamed: 0', 'PCA'])
```

```python
sample_sizes
```

```python
for i in range(len(sample_sizes)):
    sample_size = sample_sizes[i]
    fig, ax = plt.subplots()
    # Create lists for the plot
    Methods = [ 'UMAP', 'tSNE', '2 Layer AE', '6 Layer AE', 'VAE']
    x_pos = np.arange(len(Methods))
    #ax.bar(x_pos, np.array(Means[0:1])[0],  align='center', alpha=0.5, ecolor='black', capsize=10)

    ax.bar(x_pos, np.array(Means[i:i+1])[0], yerr=np.array(SDs[i:i+1])[0], align='center', alpha=0.5, ecolor='black', capsize=10)
    ax.set_ylabel('Average Absolute Correlation Coefficient')
    ax.set_xlabel('Dimensionality Reduction Method')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(Methods)
    plt.title("Samusik01 " + str(sample_size) + " Cell Subsamples", fontweight = 'bold')
    ax.yaxis.grid(True)
    ax.set_ylim([0.0, 1.0])
    # Save the figure and show
    plt.tight_layout()
    plt.savefig('../Figures/s01_PearsonSubsample' + str(sample_size) + '.png')
    plt.show()

```

```python
fig = plt.figure()
plt.xscale('log')
#plt.xscale('log')

x = sample_sizes
plt.errorbar(x, Means["UMAP"], yerr = SDs["UMAP"], label = "UMAP")
plt.errorbar(x, Means["tSNE"], yerr = SDs["tSNE"], label = "tSNE")
plt.errorbar(x, Means["AE2layer"], yerr = SDs["AE2layer"], label = "AE2Layer")
plt.errorbar(x, Means["AE6layer"], yerr = SDs["AE6layer"], label = "AE6Layer")
plt.errorbar(x, Means["VAE"], yerr = SDs["VAE"], label = "VAE")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Average Unsigned Pearson Correlation Coefficient for Samusik 01\nOf The Points' Coordinates in the Embedding of Subsamples\n Versus in the Embedding of the Full Samusik 01 Dataset", fontweight = 'bold')
plt.ylabel("Average Absolute Correlation Coefficient")
plt.xlabel("Number of Input Data Points (Subsample Size)")
plt.savefig("../Figures/PearsonSubsampleLinePlot_s01.png", bbox_inches = 'tight')
```
