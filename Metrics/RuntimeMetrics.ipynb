```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Means = pd.read_csv("Mean_Runtimes.csv")
SDs = pd.read_csv("SD_Runtimes.csv")
```

```python
sample_sizes = np.array(SDs["Unnamed: 0"])
SDs.drop(columns=['Unnamed: 0'])
Means.drop(columns=['Unnamed: 0'])
```

```python
fig = plt.figure()
plt.yscale('log')
#plt.xscale('log')

x = sample_sizes
plt.errorbar(x, Means["UMAP"], yerr = SDs["UMAP"], label = "UMAP")
plt.errorbar(x, Means["tSNE"], yerr = SDs["tSNE"], label = "tSNE")
plt.errorbar(x, Means["PCA"], yerr = SDs["PCA"], label = "PCA")
plt.errorbar(x, Means["AE2layer"], yerr = SDs["AE2layer"], label = "AE2Layer")
plt.errorbar(x, Means["AE6layer"], yerr = SDs["AE6layer"], label = "AE6Layer")
plt.errorbar(x, Means["VAE"], yerr = SDs["VAE"], label = "VAE")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Run Times of Dimensionality Reduction Methods\n For Inputs of Varying Sizes")
plt.ylabel("Average time (s)")
plt.xlabel("Number of input data points")
plt.savefig("../Figures/Runtimes.png", bbox_inches = 'tight')
```
