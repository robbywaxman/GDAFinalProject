Import Statements:

```python
import sys
import os

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import FlowCal

from umap import UMAP
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

np.random.seed(1)

import random
from scipy.stats import pearsonr
import time

import keras
from keras.utils import Sequence
from keras.layers import Input, Dense, Lambda, Layer, Add, Multiply, PReLU
from keras.models import Model, Sequential

from keras import backend as K
import tensorflow as tf
tf.random.set_seed(1)
```

#### Load in Samusik Data and the full embeddings

```python
data_dir = "../SamusikData/"
s01_fn = data_dir + "BM2_cct_normalized_01_non-Neutrophils.fcs"
s01 = FlowCal.io.FCSData(s01_fn)

# Only relevant indexes are 10-47
s01 = s01[:, 10:47]

# Transform data with arcsinh
s01 = np.arcsinh(s01)
s01_umap = np.load("../Embeddings/s01_umap.npy")
s01_tsne = np.load("../Embeddings/s01_tsne.npy")
s01_pca = np.load("../Embeddings/s01_pca.npy")
s01_2layer = np.load("../Embeddings/s01_2layer.npy")
s01_6layer = np.load("../Embeddings/s01_6layer.npy")
s01_vae = np.load("../Embeddings/s01_vae.npy")
```

#### 2 Layer AE Model:

```python
AE2layer = Sequential()
AE2layer.add(Dense(2, input_dim=37, activation="selu", name="encoder"))
AE2layer.add(Dense(37, name="decoder"))
AE2layer.compile(loss="mse", optimizer="adam", metrics=["mse"])
```

#### 6 Layer AE Model

```python
AE6layer = Sequential()
AE6layer.add(Dense(20, input_dim=37, activation="selu", name="encoder1"))
AE6layer.add(Dense(10, activation="selu", name="encoder2"))
AE6layer.add(Dense(2, activation="selu", name="encoder3"))
AE6layer.add(Dense(10, activation="selu", name="decoder1"))
AE6layer.add(Dense(20, activation="selu", name="decoder2"))
AE6layer.add(Dense(37, name="decoder3"))
AE6layer.compile(loss="mse", optimizer="adam", metrics=["mse"])
```

### VAE Model

```python
def rec_loss(y_true, y_pred):
    """ Reconstruction error """

    return K.sum(K.square(y_true - y_pred))


class KLDivergenceLayer(Layer):

    """ Identity transform layer that adds KL divergence
    to the final model loss.
    """

    def __init__(self, *args, **kwargs):
        self.is_placeholder = True
        super(KLDivergenceLayer, self).__init__(*args, **kwargs)

    def call(self, inputs):

        mu, log_var = inputs

        kl_batch = - .5 * K.sum(1 + log_var -
                                K.square(mu) -
                                K.exp(log_var), axis=-1)

        self.add_loss(K.mean(kl_batch), inputs=inputs)

        return inputs
```

```python
stdev = 1
    
x = Input(shape=(37,), name="input")
E1 = Dense(20, activation="selu", name="E1")(x)
E2 = Dense(10, activation="selu", name="E2")(E1)
mu = Dense(2, name="mu")(E2)
log_var = Dense(2, name="log_var")(E2)

KL = KLDivergenceLayer(name="KL")([mu, log_var])
sigma = Lambda(lambda t: K.exp(.5*t), name="sigma")(log_var)

z_eps = Multiply(name="z_eps")([sigma, K.random_normal(stddev=stdev,shape=(K.shape(x)[0], 2))])
z = Add(name="z")([mu, z_eps]) # not zeps

D1 = Dense(10, activation="selu", name="D1")(z)
D2 = Dense(20, activation="selu", name="D2")(D1)
D3 = Dense(37, name="D3")(D2)

VAE = Model(inputs=x, outputs=D3)
VAE.compile(loss=rec_loss, optimizer="adam", metrics=["mse"])
```

#### List of Dimensionality Reduction Methods to be Tested

```python
Methods = [
    UMAP(n_neighbors=15,min_dist=0.2,metric="euclidean",verbose=True,random_state=123),
    TSNE(n_iter=1000,verbose=True),
    PCA(),
    AE2layer,
    AE6layer,
    VAE
        ]
```

#### List of Full Embeddings to be compared to

```python
Full_s01_Embeddings = [
    s01_umap,
    s01_tsne,
    s01_pca,
    s01_2layer,
    s01_6layer,
    s01_vae   
        ]
```

#### List of Sample Sizes to be Tested

```python
Sample_Sizes = [100,200,500,1000,2000,5000,10000,20000,50000]
```

#### Lists to Store info on Pearson r and runtimes (mean and standard dev)

```python
Mean_Pearsons = [list() for x in range(len(Methods))]
SD_Pearsons = [list() for x in range(len(Methods))]
Mean_Runtimes = [list() for x in range(len(Methods))]
SD_Runtimes = [list() for x in range(len(Methods))]
```

#### Loop through and run the different methods on the different subsample sizes 3x each and compute runtime and pearson correlation for every method and sample size combination

```python
for method_idx in range(len(Methods)):
    method = Methods[method_idx]
    s01_full_embedding = Full_s01_Embeddings[method_idx]
    mean_pearson_list = Mean_Pearsons[method_idx]
    sd_pearson_list = SD_Pearsons[method_idx]
    runtime_list = Mean_Runtimes[method_idx]
    sd_runtime_list = SD_Runtimes[method_idx]
    for sample_size in Sample_Sizes:
        temp_pearson_list = []
        temp_runtime_list = []
        for i in range(3): #Repeat 3 times for each method & sample size
            indices = random.sample(range(s01.shape[0]), sample_size)
            s01_sub = s01[indices]
            s01_full_embedding_sub = s01_full_embedding[indices]
            t = time.process_time()
            
            if method == AE2layer:
                history_2layer = AE2layer.fit(s01_sub, s01_sub, 
                              use_multiprocessing=True,
                              workers=-1,
                              epochs=10)
                simple_encoder = AE2layer.get_layer("encoder")
                s01_sub_embedding = simple_encoder(s01_sub)
                
            elif method == AE6layer:
                history_6layer = AE6layer.fit(s01_sub, s01_sub,
                              use_multiprocessing=True,
                              workers=-1,
                              epochs=20)
                e1 = AE6layer.get_layer("encoder1")
                e2 = AE6layer.get_layer("encoder2")
                e3 = AE6layer.get_layer("encoder3")
                s01_sub_embedding= e3(e2(e1(s01_sub)))
                
            elif method == VAE:
                VAE_history = VAE.fit(s01_sub, s01_sub,
                              use_multiprocessing=True,
                              workers=-1,
                              epochs=20,
                              shuffle=True)
                e1 = VAE.get_layer("E1")
                e2 = VAE.get_layer("E2")
                e3 = VAE.get_layer("mu")
                s01_sub_embedding = e3(e2(e1(s01_sub)))
                
            else:
                s01_sub_embedding = method.fit_transform(s01_sub)
                
            elapsed_time = time.process_time() - t
            r_x, _ = pearsonr(s01_full_embedding_sub[:,0], s01_sub_embedding[:,0])
            r_y, _ = pearsonr(s01_full_embedding_sub[:,1], s01_sub_embedding[:,1])
            temp_pearson_list.append((abs(r_x) +abs(r_y))/2)
            temp_runtime_list.append(elapsed_time)
        mean_pearson_list.append(np.mean(np.array(temp_pearson_list)))
        sd_pearson_list.append(np.std(np.array(temp_pearson_list)))
        runtime_list.append(np.mean(np.array(temp_runtime_list)))
        sd_runtime_list.append(np.std(np.array(temp_runtime_list)))
                
```
