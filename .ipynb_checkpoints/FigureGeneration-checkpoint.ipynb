## Code to generate figures of tSNE and UMAP for Samusik01

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import copy
import matplotlib.colors
```

#### Import the data from previously run UMAP and tSNE, as wel as read in the population assignments (for coloring)

```python
s01_umap = np.load("s01_umap.npy")
s01_tsne = np.load("s01_tsne.npy")
PopData = pd.read_csv("SamusikData/population_assignments01.csv")
Assignments = np.array(PopData.Population)
```

#### Here I define my own color_map - feel free to change the colors as you wish. Since 0 is unlabeled event I put it as beige

```python
color_list = ["beige","violet","red","blue","lightsalmon", "cyan","orchid", "plum", 'olivedrab', 'palegreen', 'gold', 'tan', 'slategray','lemonchiffon', "paleturquoise","pink","crimson","orange","mediumvioletred","lightskyblue","navajowhite","sienna","springgreen","c","purple"]
cmap_own = matplotlib.colors.LinearSegmentedColormap.from_list("", color_list)
```

### UMAP Plotting

```python
plt.scatter(s01_umap[:, 0], s01_umap[:, 1], c = Assignments,s=5, alpha=0.4, cmap = cmap_own)
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.title("UMAP1 v UMAP2 for Samusik01")
plt.savefig("Figures/s01_umap.png")
plt.show()
```

### tSNE Plotting

```python
plt.scatter(s01_tsne[:, 0], s01_tsne[:, 1], c = Assignments,s=5, alpha=0.4, cmap = cmap_own)
plt.xlabel("tSNE1")
plt.ylabel("tSNE2")
plt.title("tSNE1 v tSNE2 for Samusik01")
plt.savefig("Figures/s01_tsne.png")
plt.show()
```
